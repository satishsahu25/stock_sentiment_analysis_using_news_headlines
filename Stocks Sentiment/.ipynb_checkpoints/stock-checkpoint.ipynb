```python
import pandas as pd
```

```python
# it is for a company has 25 headlines
df=pd.read_csv('data.csv',encoding='ISO-8859-1')
# encoding handles special characters
df.head(-5)
# 0 means - impact 1 means +ve imapct
```

```python
# data column is not neccessary
train=df[df['Date']<'20140101']
test=df[df['Date']>'20081231']
```

```python
test
```

```python
# removing the pucntuation/special charcaters as not required
data=train.iloc[:,2:27] #taking all rows and column from 2 to 27
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
# renaming the column names

list1=[i for i in range(25)]
newindex=[str(i) for i in list1]
data.columns=newindex
data.head(5)
```

```python
# converting headlines into lowercase by str.lower()
for index in newindex:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
# converting each row into a single sentence
# Example how to do it for one row
' '.join(str(x) for x in data.iloc[1,0:25])

# that is making a paragraph from 25 headlines
```

```python
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))

# here merging all the headliens of the all column into a single para
```

```python
headlines[2]
```

```python
# countvectorizer or tfid
from sklearn.feature_extraction.text import CountVectorizer

# countvectorizer converts the sentence into a vector( or bag of words)

from sklearn.ensemble import RandomForestClassifier
```

```python
# implement of bag of words
countvector=CountVectorizer(ngram_range=(2,2)) #(2,2) means each word will be 
# combination of two words like (the cat) (is dancing) ke kiche 0 1 if presenece
traindataset=countvector.fit_transform(headlines)
```

```python
traindataset[0]
# traindataset is sparse matrix as combiantion of 0/1 
```

```python
# randomforestclassifer
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# predicting for the testd dataset
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))

testdataset=countvector.transform(test_transform)
predictions=randomclassifier.predict(testdataset)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

# BY TFID 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
tfidvector=TfidfVectorizer(ngram_range=(2,2)) 
traindataset=tfidvector.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# predicting for the testd dataset
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))

testdataset=tfidvector.transform(test_transform)
predictions=randomclassifier.predict(testdataset)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

```python
# BY NAIVE BAYES classifier
from sklearn.naive_bayes import MultinomialNB
```

```python
naive=MultinomialNB()
naive.fit(traindataset,train['Label'])
```

```python
predictednavie=naive.predict(testdataset)
```

```python
matrix=confusion_matrix(test['Label'],predictednavie)
print(matrix)
score=accuracy_score(test['Label'],predictednavie)
print(score)
report=classification_report(test['Label'],predictednavie)
print(report)
```
